```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')

# Computer the mean and standard deviation of the data
print (df['m'].mean())
print (df['m'].std())


# Scale the mileage data
df['m'] = (df['m'] - df['m'].mean()) / df['m'].std()

# Plot the data
plt.scatter(df['m'], df['price'])
plt.xlabel('Mileage (scaled)')
plt.ylabel('Price')
plt.show()
```
