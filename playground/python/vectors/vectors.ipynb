```python
import numpy as np
import matplotlib.pyplot as plt

# Define two orthogonal vectors
u = np.array([1, 0])  # Vector u along x-axis
v = np.array([0, 1])  # Vector v along y-axis

# Calculate their dot product
dot_product = np.dot(u, v)

# Visualize the vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector v')

# Set plot limits
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Add labels and legend
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f"Visualization of Two Orthogonal Vectors (Dot Product = {dot_product})")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
