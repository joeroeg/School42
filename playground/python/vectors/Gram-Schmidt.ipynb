```python
import numpy as np
import matplotlib.pyplot as plt

def gram_schmidt(vectors):
    orthogonal_vectors = []
    for v in vectors:
        proj = sum(np.dot(v, u) / np.dot(u, u) * u for u in orthogonal_vectors)
        orthogonal_vectors.append(v - proj)
    return np.array(orthogonal_vectors)

# Original set of linearly independent vectors
v1 = np.array([2, 3])
v2 = np.array([4, 1])

# Apply Gram-Schmidt process
vectors = np.array([v1, v2])
orthogonal_vectors = gram_schmidt(vectors)

# Normalize the orthogonal vectors (optional)
orthonormal_vectors = np.array([v / np.linalg.norm(v) for v in orthogonal_vectors])

# Visualization
plt.figure(figsize=(8, 8))
origin = np.zeros(2)

# Original vectors
plt.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='blue', label='v1 (original)')
plt.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='red', label='v2 (original)')

# Orthogonal vectors
# plt.quiver(*origin, *orthogonal_vectors[0], angles='xy', scale_units='xy', scale=1, color='green', label='u1 (orthogonal)')
# plt.quiver(*origin, *orthogonal_vectors[1], angles='xy', scale_units='xy', scale=1, color='purple', label='u2 (orthogonal)')

# Set plot limits
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')

plt.title("Gram-Schmidt Process: Original and Orthogonal Vectors")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# Verify orthogonality of the resulting vectors
u1 = orthogonal_vectors[0]
u2 = orthogonal_vectors[1]

# Print vectors
print("u1:", u1)
print("u2:", u2)

# Compute dot product
dot_product_u1_u2 = np.dot(u1, u2)

dot_product_u1_u2

```

```python
# Visualization of the projection process step by step

# Original vectors
v1 = np.array([2, 3])  # u1 in the Gram-Schmidt process
v2 = np.array([4, 1])  # v2, which needs to be adjusted

# Projection of v2 onto v1
proj_v2_onto_v1 = (np.dot(v2, v1) / np.dot(v1, v1)) * v1

# Orthogonal component of v2 (u2)
u2 = v2 - proj_v2_onto_v1

# Visualization
plt.figure(figsize=(8, 8))
origin = np.zeros(2)

# Plot original vectors
plt.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color='blue', label='v1 (reference vector)')
plt.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color='red', label='v2 (original vector)')

# Plot projection of v2 onto v1
plt.quiver(*origin, *proj_v2_onto_v1, angles='xy', scale_units='xy', scale=1, color='green', label='Projection of v2 onto v1')

# Plot orthogonal component of v2 (u2)
plt.quiver(*proj_v2_onto_v1, *(u2), angles='xy', scale_units='xy', scale=1, color='purple', label='Orthogonal component (u2)')

# Set plot limits
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')

plt.title("Step-by-Step Visualization of Projection and Orthogonalization")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
